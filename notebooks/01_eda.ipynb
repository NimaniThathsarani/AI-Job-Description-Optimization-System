```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_colwidth', None)
```

```python
#Load Dataset

df = pd.read_csv("data.csv")

print(df.shape)
df.head()
```

```python
#Dataset Information
df.info()
```

```python
#Missing values
df.isnull().sum()
```

```python
#Duplicate Rows
print("Duplicates:", df.duplicated().sum())
```

```python
#Description Length Analysis
df["word_count"] = df["Description"].apply(lambda x: len(str(x).split()))

df["word_count"].describe()
```

```python
#Distribution Plot
plt.figure(figsize=(10,5))
sns.histplot(df["word_count"], bins=30)
plt.title("Distribution of Job Description Length")
plt.show()
```

```python
#Most common job titles
top_titles = df["Job Title"].value_counts().head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=top_titles.values, y=top_titles.index)
plt.title("Top Job Titles")
plt.show()
```
